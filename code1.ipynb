```python
#Data preprocessing

x,y=df.loc[:,:'thal'],df['target']
x
```

```python
y
```

```python
x.size
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=10,test_size=0.3,shuffle=True)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
#Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier

dt=DecisionTreeClassifier(max_depth=3) #maxdepth is tree depth(changes accuracy) [this func has imp parameters that affect accuracy]
dt.fit(x_train,y_train)
```

```python
prediction=dt.predict(x_test)
```

```python
prediction #prediction done on test data
```

```python
from sklearn.metrics import accuracy_score
accuracy_dect=accuracy_score(y_test,prediction)    #checking accuracy with test dataset
accuracy_dect

#you can drop columns and use max feature parameter to improve accuracy
```

```python
dt.feature_importances_
```

```python
def plot_feature_importances(model):
    plt.figure(figsize=(8,6))
    n_features = 13
    plt.barh(range(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), x)
    plt.xlabel("Feature importance")
    plt.ylabel("Feature")
    plt.ylim(-1, n_features)
plot_feature_importances(dt)
plt.savefig('feature_importance')
    
   #checks which factor(feature) affects the most for target 
```

```python
#Knn Algorithm
from sklearn.preprocessing import StandardScaler
std=StandardScaler().fit(x) 
x_std=std.transform(x)
x_std #data is scaled down (reduce difference)
```

```python
x_train_std,x_test_std,y_train,y_test=train_test_split(x_std,y,random_state=10,test_size=0.3,shuffle=True)
```

```python
x_train_std
```

```python
x_std
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_neighbors=4) #this func has parameters that affect accuracy
knn.fit(x_train_std,y_train)
```

```python
#prediction on test data
prediction_knn=knn.predict(x_test_std)
prediction_knn
```

```python
#accuracy
accuracy_knn=accuracy_score(y_test,prediction_knn)
accuracy_knn #accuracy of model
```

```python
#prediction on custom data
custom_data_knn=np.array([[68,1,0,144,193,1,1,141,0,3.4,1,2,3]])
custom_data_knn_std=StandardScaler().fit(custom_data_knn) 
custom_data_knn_std=std.transform(custom_data_knn)    #custom data also must be scaled(Standard scaler)
custom_data_knn_std
```

```python
prediction_cust_knn=knn.predict(custom_data_knn_std)
categoryknn=['No you dont  have heart disease','yes you have heart disease,sorry bro']
print(categoryknn[int(prediction_cust_knn)])
```

```python
k_range=range(1,26)  #check accuracy score for various k neighbor parameter values
scores={}
scores_list=[]
k=0
for k in k_range:
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train_std,y_train)
    pred_knn=knn.predict(x_test_std)
    scores[k]=accuracy_score(y_test,pred_knn)
    scores_list.append(accuracy_score(y_test,prediction_knn))
```
